```python
import pandas as pd
```

```python
data=pd.read_csv('Datasets.csv')
```

```python
data.drop(columns=['FN','Active'], inplace=True)
```

```python
data[data['age'].isna()]
```

```python
data['age'].fillna(data['age'].median(), inplace=True)
```

```python
data[data['age'].isna()]
```

```python
data[data['club_member_status'].isna()]
```

```python
data['club_member_status'].fillna('ACTIVE', inplace=True)
```

```python
data[data['club_member_status'].isna()]
```

```python
data[data['fashion_news_frequency'].isna()]
```

```python
data['fashion_news_frequency'].fillna(method ='ffill', inplace=True)
```

```python
data[data['fashion_news_frequency'].isna()]
```

```python
data[data['detail_desc'].isna()]
```

```python
data['detail_desc'].fillna(method ='ffill', inplace=True)
```

```python
data[data['detail_desc'].isna()]
```

```python
data.sale
```

```python
data['sale'].replace(['No', 'yes'],
                        [0, 1], inplace=True)
```

```python
data[data['sale']==0]
```

```python
data = data.drop_duplicates(keep='first')
```

```python
data.duplicated().sum()
```

```python
data.to_csv('Dataset_cleaned.csv', index=False)
```
