```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df1 = pd.read_csv("articles.csv")
```

```python
df2 = pd.read_csv("customers.csv")
```

```python
df3 = pd.read_csv("test_data.csv")
```

```python
df4 = df3.merge(df1, on='article_id', how='left').merge(df2, on='customer_id', how='left')
```

```python
df4.drop(columns=['FN','Active'], inplace=True)
```

```python
df4[df4['age'].isna()]
```

```python
df4['age'].fillna(df4['age'].median(), inplace=True)
df4[df4['age'].isna()]
```

```python
df4[df4['club_member_status'].isna()]
```

```python
df4['club_member_status'].fillna('ACTIVE', inplace=True)
df4[df4['club_member_status'].isna()]
```

```python
df4[df4['fashion_news_frequency'].isna()]
```

```python
df4['fashion_news_frequency'].fillna(method ='ffill', inplace=True)
df4[df4['fashion_news_frequency'].isna()]
```

```python
df4[df4['detail_desc'].isna()]
```

```python
df4['detail_desc'].fillna(method ='ffill', inplace=True)
df4[df4['detail_desc'].isna()]
```
